```python
import pandas as pd
import emoji
```

```python
my_df = pd.read_csv(r'/content/balanced_validation_En.csv')
```

```python
# Remove emoji from train data

for index, row in my_df.iterrows():
  my_df.at[index, 'tweet'] = emoji.get_emoji_regexp().sub(r'', row['tweet'])
```

```python
my_df
```

```python
result = my_df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'rephrase', 'sarcasm', 'irony', 'satire', 'understatement', 'overstatement', 'rhetorical_question'])
```

```python
result2 = result.rename(columns={"tweet": "texts", "sarcastic": "info"})
```

```python
result2
```

```python
info_dict_list = [{'label': row['info']} for index, row in result2.iterrows()]
```

```python
result2['info'] = info_dict_list
```

```python
result2
```

```python
result2.to_pickle("output.pkl")  
```
